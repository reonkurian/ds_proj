```python
data = pd.read_csv('../input/google-play-store-apps/googleplaystore.csv')
```

```python
# Eliminate some characters to transform Installs to numeric
# Eliminate M to transform Size to numeric
# Replace cells with k to 0 since it is < 1MB
 # Transform reviews to numeric
# Remove currency symbol from Price, change it to numeric
# Last Updated to date format
# Replace "Varies with device" to NA since it is unknown
# Keep only version number to 1 decimal
# Two apps had type as 0 or NA, they will be removed 
```

```python
data['installs_cleaned']=data['Installs'].apply(lambda x:x.split('+')[0])
```

```python
data['installs_cleaned']=data['installs_cleaned'].apply(lambda x:x.replace(',',''))
```

```python
data = data.drop(data[data.installs_cleaned == 'Free'].index)

```

```python
data["installs_cleaned"] = pd.to_numeric(data["installs_cleaned"])
```

```python
data['installs_cleaned'].dtype
```

```python
# Eliminate M to transform Size to numeric
# Replace cells with k to 0 since it is < 1MB
```

```python
data['size_cleaned']=data['Size'].apply(lambda x: x.replace('M',''))
```

```python
data= data.drop(data[data.size_cleaned =='Varies with device'].index)
```

```python
ending = ['k']
data['size_kb']=data['size_cleaned'].map(lambda x: x[-1:] in ending) 
```

```python
data['size_cleaned']=data['size_cleaned'].apply(lambda x: x.replace('k',''))
```

```python
data['size_cleaned']=pd.to_numeric(data['size_cleaned'])
data['size_cleaned'].dtype
```

```python
data['size_cleaned']=data.apply(lambda x:x['size_cleaned']/1000 if x['size_kb']==1 else x.size_cleaned,axis=1)    
```
