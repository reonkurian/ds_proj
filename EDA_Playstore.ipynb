```python
df_cat = df[['Price','Installs','Reviews','Category','Current Ver','Type','Content Rating','Genres']]
```

```python
for i in df_cat[['Current Ver','Installs','Reviews','Category','Type','Content Rating','Genres']].columns:
    cat_num = df_cat[i].value_counts()[:20]
    print("graph for %s: total = %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df[['Rating','Reviews','installs_cleaned','Price','size_cleaned']].corr(),vmax=.3, center=0, cmap=cmap,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
g = sns.catplot(x="Category",y="Rating",data=df, kind="box", height = 10 ,
palette = "Set1")
g.despine(left=True)
g.set_xticklabels(rotation=90)
g.set( xticks=range(0,34))
g = g.set_ylabels("Rating")
plt.title('Boxplot of Rating VS Category',size = 20)
```

```python
plt.figure(figsize = (10,10))
sns.regplot(x="Reviews", y="Rating", color = 'darkorange',data=df[df['Reviews']<1000000]);
plt.title('Rating VS Reveiws',size = 20)
```

```python
plt.figure(figsize = (10,10))
sns.regplot(x="Price", y="Rating", color = 'darkorange',data=df[df['Price']<80]);
plt.title('Price VS Rating',size = 20)
```

```python
plt.figure(figsize = (10,10))
sns.regplot(x="size_cleaned", y="Rating", color = 'darkorange',data=df);
plt.title('Size VS Rating',size = 20)
```
